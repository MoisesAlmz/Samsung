# Actividad: Operaciones con Data Frames


## Instrucciones


1. Dado el siguiente diccionario construya un objeto DataFrame.
2. Filtre los resultados por año de publicación.
3. Obtenga algunas columnas basado en que contenga alguna palabra en el nombre de la publicacion.


```python

import pandas as pd
publicaciones_literarias = {
    1605: "Don Quijote de la Mancha - Miguel de Cervantes",
    1865: "Alicia en el país de las maravillas - Lewis Carroll",
    1949: "1984 - George Orwell",
    1954: "El Señor de los Anillos - J.R.R. Tolkien",
    1997: "Harry Potter y la piedra filosofal - J.K. Rowling",
    2003: "El Código Da Vinci - Dan Brown"
}
# Espacio para desarrollar el código y agregar los comentarios necesarios.

print("Objeto data frame: ")
obj = list(publicaciones_literarias.items())
data_frame_libros = pd.DataFrame(obj, columns=["Año", "Publicación"])
print(data_frame_libros)

print("\n")
print("Filtrado de los resultados por año de publicación:")

data_frame_1605 = data_frame_libros[data_frame_libros["Año"] == 1605]
print(data_frame_1605)

data_frame_1865 = data_frame_libros[data_frame_libros["Año"] == 1865]
print(data_frame_1865)

data_frame_1949 = data_frame_libros[data_frame_libros["Año"] == 1949]
print(data_frame_1949)

data_frame_1954 = data_frame_libros[data_frame_libros["Año"] == 1954]
print(data_frame_1954)

data_frame_1997 = data_frame_libros[data_frame_libros["Año"] == 1997]
print(data_frame_1997)

data_frame_2003 = data_frame_libros[data_frame_libros["Año"] == 2003]
print(data_frame_2003)

print("\n")

print("Columnas basadas en que contenga alguna palabra en el nombre de la publicacion:")

data_frame_Quijote = data_frame_libros[data_frame_libros["Publicación"].str.contains("Quijote")]
print(data_frame_Quijote)
```
